```python
import numpy as np
```

```python
states = [0, 1, 2, 3]
actions = ['left', 'right']
transition_probabilities = {
    0: {'left': [(0.8, 0), (0.2, 1)], 'right': [(1.0, 1)]},
    1: {'left': [(0.7, 0), (0.3, 2)], 'right': [(0.8, 2), (0.2, 1)]},
    2: {'left': [(0.6, 1), (0.4, 3)], 'right': [(0.9, 3), (0.1, 2)]},
    3: {'left': [(1.0, 3)], 'right': [(1.0, 3)]}
}
rewards = {
    0: {'left': 0, 'right': 0},
    1: {'left': 0, 'right': 0},
    2: {'left': 0, 'right': 1},
    3: {'left': 0, 'right': 0}
}
gamma = 0.9
```

```python
V = np.zeros(len(states))
policy = {s: 'left' for s in states}
```

```python
def policy_evaluation(policy, V, threshold=1e-6):
    while True:
        delta = 0
        for s in states:
            a = policy[s]
            v = V[s]
            V[s] = sum(p * (rewards[s][a] + gamma * V[s_]) for p, s_ in transition_probabilities[s][a])
            delta = max(delta, abs(v - V[s]))
        if delta < threshold:
            break
```

```python
def policy_improvement(V):
    policy_stable = True
    for s in states:
        old_action = policy[s]
        action_values = {}
        for a in actions:
            action_values[a] = sum(p * (rewards[s][a] + gamma * V[s_]) for p, s_ in transition_probabilities[s][a])
        policy[s] = max(action_values, key=action_values.get)
        if old_action != policy[s]:
            policy_stable = False
    return policy_stable
```

```python
while True:
    policy_evaluation(policy, V)
    if policy_improvement(V):
        break
```

```python
print("Optimal Policy:", policy)
print("Value Function:", V)
```
